```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
df=pd.read_csv('Toyota.csv')
```

```python
df.head()
```

```python
df.drop('Car_Company',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
model_names=list(df['Model_Name'].unique())
```

```python
len(model_names)
```

```python
avg_prices=[]
for i in model_names:
    avg_prices.append(df[df['Model_Name']==i]['selling_price'].mean())
```

```python
models_df=pd.DataFrame({'Model_name':model_names,'Avg_selling_price':avg_prices})
```

```python
models_df
```

```python
models_dict={}
```

```python
for i in range(0,88):
    models_dict.update([(model_names[i],avg_prices[i])])
```

```python
len(models_dict)
```

```python
Newf_df=df.copy()
```

```python
Newf_df.head()
```

```python
Newf_df.info()
```

```python
Newf_df['Model_Name']=Newf_df['Model_Name'].map(models_dict)
```

```python
df['fuel'].unique()
```

```python
fuel_input={'Petrol':0,'Diesel':1,'Electric':2}
Newf_df['fuel']=Newf_df['fuel'].map(fuel_input)
```

```python
seller_input={'Individual':0,'Dealer':1,'Trustmark Dealer':2}

Newf_df['seller_type']=Newf_df['seller_type'].map(seller_input)
```

```python
trans_input={'Manual':0,'Automatic':1}
    
Newf_df['transmission']=Newf_df['transmission'].map(trans_input)
```

```python
owner_input={'First Owner':0,'Second Owner':1,'Third Owner':2,'Fourth & Above Owner':3}

Newf_df['owner']=Newf_df['owner'].map(owner_input)
```

```python
Newf_df
```

```python
y=Newf_df['selling_price']
```

```python
Newf_df.drop('selling_price',axis=1,inplace=True)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
random_params={'n_estimators':[100,110,120,130],'max_depth':[9,10,11,12,13],'min_samples_leaf':[1,2,3]}
```

```python
random_grid=GridSearchCV(estimator=RandomForestRegressor(),param_grid=random_params,cv=5,verbose=3)
```

```python
random_grid.fit(Newf_df,y)
```

```python
random_grid.best_score_
```

```python
random_grid.best_params_
```

```python
Toyota_model=RandomForestRegressor(n_estimators=130,max_depth=12,min_samples_leaf=2)
```

```python
Toyota_model.fit(Newf_df,y)
```

```python
pickle.dump(Toyota_model,open('Toyota_model.pkl','wb'))
```
